```python
from sqlson import SQLDict
# from sqldict import SQLDict as sqld2
import json
from uuid import uuid4
from secrets import token_urlsafe
# import ujson

from random import choice
from string import ascii_lowercase as letters
```

```python
def rands(r=20):
    return "".join([choice(letters) for _ in range(r)])
```

```python
dd = SQLDict("ff.db")
```

```python
%%timeit -n 2000
dd[rands()] = {
    "user": "ricardo",
    "email": "ricardo.ander@gmail.com",
    "uuid": str(uuid4()),
    "token": token_urlsafe(32),
    "rs": rands(r=100),
    "rs2": rands(r=200),
    "rs3": rands(r=300),
}
```

```python
def mp(d):
    k = d[0]
    v = d[1]
    if "a" in v["uuid"]:
        v["uuid"] = "modif"
    dd[k] = v
```

```python
%%time
list(map(mp, dd.items()))
```
